```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.arange(-10,10,0.1)
y = np.arange(-10,10,0.1)
```

```python
def z_func(x,y):
    return np.sin(np.sqrt(x**2 + y**2))
```

```python
def gradient(x, y):
    r = np.sqrt(x**2 + y**2)  # Distance from origin
    if r == 0:  # Avoid division by zero
        return 0, 0
    grad_x = np.cos(r) * (x / r)
    grad_y = np.cos(r) * (y / r)
    return grad_x, grad_y
```

```python
X ,Y = np.meshgrid(x,y)
Z = z_func(X,Y)
initial_pos = (9.9,9.9,z_func(9.9,9.9))
l_rate = 0.01
```

```python
fig = plt.figure()
ax = plt.subplot(projection= "3d")
ax.plot_surface(X,Y,Z,cmap = "viridis",alpha = 0.7)

for _ in range(1000):
    X_deriv,Y_deriv = gradient(initial_pos[0],initial_pos[1])
    X_final = initial_pos[0] - l_rate * X_deriv
    Y_final = initial_pos[1] - l_rate * Y_deriv
    Z_final = z_func(X_final, Y_final)
    initial_pos = (X_final, Y_final, Z_final)

    
    ax.scatter(initial_pos[0],initial_pos[1],initial_pos[2],color = "red",s=10)   
    plt.pause(0.01)

plt.show()
```
