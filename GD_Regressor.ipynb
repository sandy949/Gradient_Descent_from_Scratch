```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Set random seed for reproducibility
np.random.seed(42)

# Generate input variables
X1 = np.random.rand(1000)  # Random values between 0 and 1
X2 = np.random.rand(1000)  # Random values between 0 and 1

Y = np.sin(np.sqrt(X1**2 + X2**2)) + np.random.rand(1000)  

dataset = pd.DataFrame({'X1' : X1,
                        'X2' : X2,
                        'Y' : Y})
```

```python
# Plotting the noisy 3D scatter plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of X1, X2, and Y
ax.scatter(X1, X2, Y, c=Y, cmap='viridis')

# Labels and title
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')
ax.set_title('3D Scatter Plot of Noisy Function')

# Show plot
plt.show()
```
